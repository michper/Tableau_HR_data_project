```python
#BIBLIOTEKI
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import openpyxl
```

```python
#USTAWIENIA
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)
pd.set_option('display.colheader_justify', 'center')
pd.set_option('display.precision', 2)
```

```python
# connected to local file
path_to_file = r'C:\Praca_Projekty\Tableau_data\HRDataset_source.csv'
df = pd.read_csv(path_to_file)
```

```python
# mapping for GenderID and MarriedID
gender_mapping = {
    0: 'Female',
    1: 'Male'
}

married_mapping = {
    0: 'No',
    1: 'Yes'
}

columns_to_be_mapped = {'GenderID': gender_mapping, 'MarriedID': married_mapping}

for column in columns_to_be_mapped:
    df[column+"_mapped"] = df[column].map(columns_to_be_mapped[column]) 
```

```python
# pull the year from DOB and show the distribution on chart
df['DOB_year'] = '19' + df['DOB'].str[6:]

df.groupby(by='DOB_year')['DOB_year'].count().plot();
df.groupby(by='DOB_year')['DOB_year'].size().reset_index(name='Count') \
                                     .sort_values(by='Count', ascending=False).set_index('DOB_year')
```

```python
gender_distribution = df['GenderID_mapped'].value_counts()

fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(aspect="equal"))
label = ['Female','Male']

plt.pie(gender_distribution, labels=label,
        autopct='%1.1f%%', pctdistance=0.7)
circle = plt.Circle( (0,0), 0.5, color='white')
p=plt.gcf()
p.gca().add_artist(circle)
ax.set_title("Gender distribution",fontdict={'fontsize': 15})
plt.tight_layout()
plt.show()

df['GenderID_mapped'].value_counts()
```

```python
# create df with checking mean salary for grouped EmpSatisffaction columns
df_grouped = df[['EmpSatisfaction','GenderID_mapped','Salary']].groupby(by=['EmpSatisfaction','GenderID_mapped']).mean()
```

```python
# create seaborn chart
min_value = df['EmpSatisfaction'].min() # find min satisfaction
max_value = df['EmpSatisfaction'].max() # find max satisfaction

plt.figure(figsize=(15,5)) # set size of chart
g = sns.lineplot(data=df_grouped, x='EmpSatisfaction', y='Salary', hue='GenderID_mapped', ci=None) # create chart
plt.xticks(np.arange(min_value, max_value+1, 1.0)) # set the frequency on X axis
plt.legend(bbox_to_anchor=(1.01, 1), loc='upper left', borderaxespad=0); # place the legend

df_grouped
```

```python
# create buckets for Salary
max_value = df['Salary'].max()+10 # find max salary +10 for below rounding calculation to do not exclude 1 record
min_value = df['Salary'].min()-10 # find min salary -10 for below rounding calculation to do not exclude 1 record
number_of_buckets = 10 # choose the number of buckets

buckets_for_salary = range(min_value,max_value, round((max_value-min_value)/number_of_buckets)) # create buckets 

# add new column with buckets mapping
df['Salary_buckets'] = pd.cut(df['Salary'],buckets_for_salary)
```

```python
# create grouped table and chart salary_bucket x gender
# grouped table
df_age_cp = df[['Salary_buckets','GenderID_mapped']].groupby(['Salary_buckets','GenderID_mapped']).size().unstack(fill_value=0)

# chart
fig, ax = plt.subplots()
df_age_cp_plot = df_age_cp.plot.barh(stacked=True, title='Salary by Gender', figsize=(20, 5), ax=ax)
ax.legend(["Female", "Male"])
ax.set_xlabel("Count")
ax.set_ylabel("Salary buckets");

# call out table
df_age_cp
```
